```python
import matplotlib.pyplot as plt
import numpy as np

# Data
frequency = np.array([100, 200, 300, 400, 500, 600, 700, 800, 900, 1000])
voltage = np.array([2.47, 2.47, 2.47, 2.47, 2.47, 2.47, 2.47, 2.47, 2.47, 2.47])

# Plot
plt.figure(figsize=(10, 6))
plt.plot(frequency, voltage, marker='o')
plt.xscale('log')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Voltage (V)')
plt.title('Logarithmic Graph of Frequency vs Voltage')
plt.grid(True, which="both", ls="--")
plt.show()
```
